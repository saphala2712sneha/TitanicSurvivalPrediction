# Titanic Survival Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data=pd.read_csv(r'C:\Users\SNEHA SAPHALA\Downloads\train.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.info()
```

```python
#remove missing/null values
titanic_data= titanic_data.drop(columns='Cabin', axis=1)
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python
#replace the model value with missing value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
#replacing null values with mean of 
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.describe()
```

```python
#how many survived??
titanic_data['Survived'].value_counts()
```

```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
x=titanic_data.drop(columns= ['Name', 'PassengerId', 'Ticket','Survived'])
```

```python
y=titanic_data['Survived']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.2)
```

```python
model1= LogisticRegression()
```

```python
model1.fit(x_train, y_train)
```

```python
x_train_predict= model1.predict(x_train)
x_test_predict=model1.predict(x_test)
```

```python
print(x_train_predict)
```

```python
print(x_test_predict)
```

```python
training_accuracy = accuracy_score(y_train, x_train_predict)
```

```python
print (training_accuracy)
```

```python
test_accuracy = accuracy_score(y_test, x_test_predict)
```

```python
print(test_accuracy)
```
